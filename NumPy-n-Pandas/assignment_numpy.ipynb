**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

```python
# your code
np.random.seed(42)
A = np.random.randint(0, 11, size=10)

print(A)

mask = (A > 3) & (A < 8)

A[mask] = A[mask] * -1

print(A)
```

**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
# your code
X = np.arange(2*3).reshape(2,-1) 
print(X)

col1 = np.repeat(X[0], len(X[1]))
col2 = np.tile(X[1], len(X[0]))

result = np.stack((col1, col2), axis=1)
print(result)
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

```python
# your code
A = np.random.randint(0, 5, (8, 3))
B = np.random.randint(0, 5, (2, 2))

print("A:\n", A, "\nB:\n", B)

mask = []
for a in A:
    a_unique = np.unique(a)
    mask.append(all(all(np.isin(b, a_unique)) for b in B))

result = A[mask]

print("\nMask:\n", mask)
print("\nResult:\n", result)
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

```python
# your code
A = np.random.randint(0, 5, (10, 3))

mask = A.max(axis=1) != A.min(axis=1)
result = A[mask]

print(f"Result: \n{result}")
print(f"Resault shape: \n{result.shape}")
print(f"Origin shape: \n{A.shape}")
```

* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
# your code
X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
X_list = list(X)

# Python
diag_len = min(len(X_list), len(X_list[0]))
prod = 1
for i in range(diag_len):
    if X_list[i][i] != 0:
        prod *= X[i][i]

print(prod)

# NumPy
diag = np.diagonal(X)
diag_nonzero = diag[diag != 0]
prod = np.prod(diag_nonzero)

print(prod)
```

* __Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.  
  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ True.

```python
# your code
x = np.array([1, 2, 2, 4])
y = np.array([4, 2, 1, 2])
x_list = list(x)
y_list = list(y)

# Python
same_multiset = sorted(x_list) == sorted(y_list)
print(same_multiset)

# NumPy
same_multiset = np.array_equal(np.sort(x), np.sort(y))
print(same_multiset)
```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль. 
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
# your code
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
x_list = list(x)

# Python
candidates = (x_list[i] for i in range(1, len(x_list)) if x_list[i-1] == 0)
max_value = max(candidates)
print(max_value)

# NumPy
mask = x[:-1] == 0
max_value = np.max(x[1:][mask])
print(max_value)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
# your code
x = np.array([2, 2, 2, 3, 3, 3, 5])
x_list = [2, 2, 2, 3, 3, 3, 5]

# Python
values = []
counts = []
current = x_list[0]
count = 1

for elem in x_list[1:]:
    if elem == current:
        count += 1
    else:
        values.append(current)
        counts.append(count)
        current = elem
        count = 1
values.append(current)
counts.append(count)

tup = (values, counts)
print(tup)

# NumPy
diff = np.diff(x) != 0
full_diff = np.append(diff, True)
indices = np.where(full_diff)[0]

counts = np.diff(np.append(-1, indices))
values = x[indices]

tup = (values, counts)
print(tup)
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

```python
%%time
# your code

# Python
import math

X = [[1, 2], [3, 4], [5, 6]]
Y = [[0, 0], [1, 1]]

dist_matrix = []

for x in X:
    row = []
    for y in Y:
        d = math.sqrt(sum((xi - yi)**2 for xi, yi in zip(x, y)))
        row.append(d)
    dist_matrix.append(row)

print(dist_matrix)
```

```python
%%time

# NumPy
X = np.array([[1, 2], [3, 4], [5, 6]])
Y = np.array([[0, 0], [1, 1]])

X_sq = np.sum(X**2, axis=1).reshape(-1, 1)
Y_sq = np.sum(Y**2, axis=1).reshape(1, -1)
dist_matrix = np.sqrt(X_sq + Y_sq - 2 * X @ Y.T)

print(dist_matrix)
```

```python
%%time

# SciPy
from scipy.spatial.distance import cdist

dist_matrix = cdist(X, Y, metric='euclidean')
print(dist_matrix)
```
